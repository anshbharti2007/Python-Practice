Problem 1 : age classifier !

```python

# this program will check the age and print the category of the person
age = int(input("Please Enter Your Age: "))
if age >= 0 and age <= 13:
    print("child")
elif age >= 14 and age <= 17:
    print("Teenager")
elif age >= 18 and age <= 30:
    print("adult")
else:
    print("You are too old !")
```

#Problem 2 : even or odd and sign checker

```python

# this program will check if the number is even or odd and also check if it is positive or negative
num1 = int(input("Enter a number: "))
if num1 % 2 == 0:
    print("even", end=" ")
else:
    print("odd", end=" ")
if num1 > 0:
    print ("and positive")
elif num1 < 0:
    print("and negative")
else:
    print("the number is zero")
    


```

#Problem = 4: CGPA Calculator

```python
Quiz1 =int(input("Enter Quiz1 Score: "))
Quiz2 =int(input("Enter Quiz2 Score: "))
Endterm = 60/100 * int(input("Enter Endterm Score: "))
Quiz_avg = 40/100 * (Quiz1 + Quiz2) / 2
total_score = Quiz_avg + Endterm
if total_score >= 90:
    print("Grade S")
elif total_score >= 80 and total_score < 90:
    print("Grade A")
elif total_score >= 70 and total_score < 80:
    print("Grade B")
elif total_score >= 60 and total_score < 70:
    print("Grade C")
elif total_score >= 50 and total_score < 60:   
    print("Grade D")
else:
    print("Grade F")
```

#Problem 5: Leap Year Checker !

```python

try:
    leap_year = int(input("Check if the year is Leap: "))
    if leap_year % 400 == 0:
        print("Its a Leap Year!")
    elif leap_year % 100 == 0: 
        print("Its not a Leap Year!")
    elif leap_year % 4 == 0:
        print("its a Leap Year!")
    else:
        print(" your year is not a leap year")
except ValueError:
    print("Error: Please enter the year using digits only (e.g., 2024)")
```
