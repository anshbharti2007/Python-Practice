## The `len()` and `sum()` functions 
* The `len()` function gives the number of item in the object.
* The `sum()` function calculate the total of all numerical item in iterable.

### The `len()` function 
* With **Set, List, Tuple, Dict** Return the number of item.
* It always start with `0`.
* If the dataset is empty `len()` return `0`.

```python
Example:
set1 = {"apple", 5, "bat"} # assign variable.
print(len(set1)) # print the number of item  '3'.
```

### The `sum()` function 
* **Syntax:** `sum(iterable, start =0)`
* Sums start and items of an iterable from left to right and return the total.
* The iterable is normally numbers and the start value is allowed to be a string.

```python
Example:
set1 = {5, 6, 10}  # assign variable.
print(sum(set1, 100))  # add the iterable and starts from base 100.
```
