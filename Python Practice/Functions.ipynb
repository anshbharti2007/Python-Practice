# Functions 
### `abs()`
`abs()` returns the absolute value of a number.

- For a positive number, it returns the same number.
- For a negative number, it returns the number without the minus sign.
- For zero, it returns `0`.

Examples:
- `abs(5)` → `5`
- `abs(-5)` → `5`
- `abs(0)` → `0`

It works for integers, floats, and some numeric types.
