# Functions 

### abs()
`abs()` returns the absolute value of a number.

* For a positive number, it returns the same number.
* For a negative number, it returns the number without the minus sign.
* For zero, it returns `0`.
**Examples:**

```py
abs(5) → 5
abs(-5) → 5
abs(0) → 0
```
It works for integers, floats, and some numeric types.



### `aiter()`
`aiter()` returns an asynchronous iterator for an asynchronous iterable.

* This is the asynchronous equivalent of the `iter()` function.
* It is used primarily in async programming.

```py
async_iter = aiter(async_iterable)
```
