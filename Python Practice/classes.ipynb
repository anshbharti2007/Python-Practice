```python
# Classes, Objects, and vars()

## Class
A class is a blueprint for creating objects. It groups related data (attributes) and behavior (methods) under one name. Classes define the structure and default behavior that their instances will have.

Example concept:
- Attributes: variables that hold state (e.g., name, rating).
- Methods: functions that operate on that state (e.g., serve_order()).
- Constructor (__init__): initializes instance attributes when an object is created.

## Object (Instance)
An object is a concrete realization of a class. Each object has its own set of instance attributes (its state) and can use the class methods.

Key points:
- Multiple objects can be created from a single class.
- Objects maintain their own separate state (unless attributes are defined on the class itself).

## vars()
`vars()` is a built-in helper that returns the `__dict__` attribute of an object or module — essentially a dictionary of attribute names to values.

- `vars(obj)` shows the instance attributes of `obj`.
- `vars(ClassName)` or `ClassName.__dict__` shows the class attributes and methods (including built-ins).
- Useful for quick inspection, debugging, or simple serialization.

Behavior notes:
- If an object has no instance attributes, `vars(obj)` returns an empty dict.
- `vars()` reflects only the attributes stored in `__dict__`; objects using `__slots__` may not show all attributes.

## Small example (assumes a `Restaurant` class and `bobs_burger` instance exist in the notebook)
```

```python
class Restaurant:
  name = ""
  category = ""
  rating = 0.0
  delivery = False
bobs_burger = Restaurant()
bobs_burger.name = "Bob\'s Burgers"
bobs_burger.category = "American Diner"
bobs_burger.rating = 5.0
bobs_burger.delivery = False
print(vars(bobs_burger))
```

```python
class BankAccount:
  def __init__(self, first_name, last_name, account_id, account_type, pin, balance):
    self.first_name = first_name
    self.last_name = last_name
    self.account_id = account_id
    self.account_type = account_type
    self.pin = pin
    self.balance = balance
  
  def deposit(self, amount):
    self.balance += amount
    return self.balance

  def withdraw(self, amount):
    self.balance -= amount:
    return self.balance

  def display_balance(self):
    print(f"{self.first_name} {self.last_name}'s balance is {self.balance}")
```

```python
# Short Pokedex with 3 Pokemon
class Pokemon:
  def __init__(self, entry, name, types, description, is_caught):
    self.entry = entry
    self.name = name 
    self.types = types
    self.description = description 
    self.is_caught = is_caught
  
  def speak(self):
    print(f"{self.name}....")
  def display_details(self):
    print(
    f"Entry Number: {self.entry}\n"
    f"Name: {self.name}\n"
    f"Type: {self.types}\n"
    f"Description: {self.description}\n"
    f"{self.name} has already been caught!\n"
    )

Pokemon1 = Pokemon(25, "Pickachu", ["Electric"], "Pickachu is a yellow mouse pokemon", "False")
Pokemon2 = Pokemon(143, "Charlizard", ["Fire", "Flying"], "Charlizard is a dragon pokemon but not a dragon lol", "True")
Pokemon3 = Pokemon(6, "Snorlax", ["Normal"], "Snorlax is a big life size teddy pokemon like eating and sleeping", "True")
Pokemon1.speak()
Pokemon1.display_details()
Pokemon2.speak()
Pokemon2.display_details()
Pokemon3.speak()
Pokemon3.display_details()
```
