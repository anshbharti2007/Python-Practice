```python
import imageio.v3 as iio
import os

# 1. Set the directory where your images are located
image_dir = "/home/shinon/Documents/GitHub/Python-Practice/PythonPractice/"

# 2. List the filenames
filenames = ["1.png", "2.png", "3.png", "4.png"]

images = []

for filename in filenames:
    # os.path.join ensures the path is correctly built for Linux
    full_path = os.path.join(image_dir, filename)
    print(f"Reading: {full_path}")
    images.append(iio.imread(full_path))

# 3. Save the GIF into the same folder
output_path = os.path.join(image_dir, "anger_animation.gif")
iio.imwrite(output_path, images, duration=500, loop=0)

print(f"Success! GIF saved at: {output_path}")

```

os module is a module used for using your operating system a medium in coding
