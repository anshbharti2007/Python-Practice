#Practice 1 - Countdown Timer

```python
for i in range(10, 0, -1):
    print(i)
print("GO!")
    
```

#Practice 2 : sum of numbers

```python
n = int(input("Enter  a number: "))
total = 0 
for i in range(1, n+1):
    total += i
print(f"Total Sum: {total}")

```

#Practice 3 : Multiplication Table 

```python
num1 = int(input("Enter a number: "))
for i in range (1,11):
    print(f"{num1} * {i} = {num1 * i}")
```

#Practice 4: Identifying Vowels

```python
word = (input("Enter a word which you want to find the number of vowels: "))
for letter in word:
    if letter in "aeiou":
        print(letter)
```

#Practice 5: Fizz buzz 

```python
for i in range (1 , 31):
    if i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print("Buzz")
    elif i % 3 and 5 == 0: 
        print("FizzBuzz")
    else: print(i)
```
