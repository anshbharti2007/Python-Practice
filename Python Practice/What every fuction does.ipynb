# Max and Min Functions

These functions return the highest or lowest items, but they behave differently depending on whether you use numbers or strings (text).

## The `max()` Function
* **With Numbers:** Returns the highest numerical value.
* **With Strings:** Returns the letter that comes *last* in the alphabet. 
* **Note on spaces:** Spaces have the *lowest* value in programming (ASCII value 32). Because of this, `max()` will generally prioritize letters over spaces.

```python
# Example
max(10, 5, 100) # Returns 100
max("apple")    # Returns 'p'

````python

# NUMERICAL DATA
numbers_list = [10, 20, 30, 40, 50, 4 ]
numbers_tuple = (10, 20, 30, 40, 50, 4)
numbers_set = {10, 20, 30, 40, 50, 4}
numbers_dict = {1:10, 2:20, 3:30, 4:40, 5:50, 6:4}

#STRING DATA
string = "japanese"
string_list = ["English", "japanese", "Chinese", "Hindi"]
string_tuple = ("English", "japanese", "Chinese", "Hindi")
string_set = {"English", "japanese", "Chinese", "Hindi"}
string_dict = {"one":"English", "two":"japanese", "three":"Chinese", "four":"Hindi"}

#MAX FUNCTION

print(max(numbers_list)) #find the hightest numerical value
print(max(numbers_tuple)) #find the hightest numerical value
print(max(numbers_dict)) #find the hightest numerical value
print(max(numbers_set)) #find the highteset numerical value
print(max(string)) #checks the ACSII/unicode and print which one comes later (space come before any character)
print(max(string_list, )) #checks the ACSII/unicode and print which one comes later
print(max(string_tuple))  #checks the ACSII/unicode and print which one comes later
print(max(string_set))  #checks the ACSII/unicode and print which one comes later
print(max(string_dict))  #checks the ACSII/unicode and print key of which one comes later
print(max(string_dict.values()))  #checks the ACSII/unicode and print key of which one comes later

print() #Print a blank line 

#MIN FUNCTION
print(min(numbers_list)) #find the smallest numerical value
print(min(numbers_tuple)) #find the smallest numerical value
print(min(numbers_dict)) #find the smallest numerical value
print(min(numbers_set)) #find the smallest numerical value
print(min(string)) #checks the ACSII/unicode and print which one comes first (space come before any character)
print(min(string_list)) #checks the ACSII/unicode and print which one comes later
print(min(string_tuple)) #checks the ACSII/unicode and print which one comes later
print(min(string_set)) #checks the ACSII/unicode and print which one comes later
print(min(string_dict)) #checks the ACSII/unicode and print key of which one comes later
print(min(string_dict.values())) #checks the ACSII/unicode and print key of which one comes later
````
