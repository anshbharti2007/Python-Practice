# Loops

* **What is a loop?** - It's basically a way to make your code repeat itself over and over again without writing the same thing multiple times.
* In `Python`, loops help you run a block of code repeatedly until a certain condition is met.
* We say "iterate" when talking about going through each item in a loop.
* One common type is the `while` loop, which keeps going as long as the condition is true. There's also the `for` loop for iterating over lists or ranges. Let's practice with some fun examples!

### `range()`
* To loop through a set of code a specified number of times, we can use a `for` loop with the `range()` function.
* The `range()` function returns a sequence of numbers, starting from 0 by default, and incrementing by 1 up to (but not including) the specified number. 
**Example**
```python
for i in range(6):
    print(i)

The Output:
0
1
2
3
4
5
```
### Nested loops
* In this one loop (inner loop) is put inside another loop.

**Example**
```python

# Outer Loop
for i in range(1, 4): # i will be 1, 2, then 3
    print(f"Outer loop iteration: {i}")
    
    # Inner Loop
    for j in range(1, 3): # j will be 1, then 2
        print(f"  Inner loop step: {j}")

```

#Practice 6: Password Validator

````python
while True:
    password = (input("Enter Your Password: "))
    
    if len(password) < 8: 
        print("It cant be your password its too small")
        continue
    check_num = False
    for char in password:
        if char.isdigit():
            check_num = True
            break
    if not check_num:
        print("Must contain at least a number")
        continue
    print("Password set sucessfull! ")
    break
       
    
````

#Practice pyramid

````python
num_of_rows = int(input("Enter No. of row: "))
for i in range (1, num_of_rows + 1 ):
    spaces = " " * (num_of_rows - i)
    star = "*" * (2 * i - 1)
    print(spaces + star)

````

#Practice Prime number printer

````python
num1 = int(input("Enter a number: "))
if num1 < 2 : 
    print(f"{num1} is not a prime number")
else:
    is_prime = True
    for i in range(2, num1):
        if num1 % i == 0:
           is_prime = False
           break
if is_prime:
    print(f"{num1} is a prime number")
else:
    print(f"{num1} is not a prime number")
````
