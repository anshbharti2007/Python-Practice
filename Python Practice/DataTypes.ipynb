* List can have different data type values.

* List are mutable means you can change each item.

## Slicing 
* With slicing we can access certain part of datatypes 
* Instead of accessing an item using a single index like name [index], we can get multiple items by specifying where to start and where to end the range like name`[start : end]`.
```py
vowel = ["a", "e", "i" ,"o" ,"u"]
print(vowel[0:3]) #output = ['a', 'e', 'i']
print(vowel[3:5]) #output = ['o', 'u']

```
