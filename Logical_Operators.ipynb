# Logical Operators
* **Logical Operator** also knows as Boolen operators, combine and evalute two conditions. They are `and`, `or` and `not` operators:
    * `and` return True if both condition are True, and return False otherwise.
    * `or` return True if atleast one condition is True, and return 
    * `not` return True if condition is False, and vise versa.

```python
# The Cyclone roller coaster
height = float(input("What\'s your height: "))
credit = int(input("What\'s your credit: "))
if height >= 137 and credit >= 10:
  print("Enjoy the ride")
elif height < 137 and credit >= 10:
  print("You dont have enough height")
elif height >= 137 and credit < 10:
  print("You dont have enought credits")
else: print("You have not met the requirement")
```
