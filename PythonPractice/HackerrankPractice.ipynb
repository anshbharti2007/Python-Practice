# Task 1 
* The first line contains the sum of the two numbers.
* The second line contains the difference of the two numbers (first - second).
* The third line contains the product of the two numbers

```python
def calculate(a, b):
    add = (a + b)
    sub = (a - b)
    multiply = (a * b) 
    print(add)
    print(sub)
    print(multiply)

if __name__ == '__main__':
    a = int(input())
    b = int(input())
    calculate(a, b)
```
