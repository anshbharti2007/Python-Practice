## Slicing 
* With slicing we can access certain part of datatypes 
* Instead of accessing an item using a single index like name [index], we can get multiple items by specifying where to start and where to end the range like name`[start : end]`.
```py
vowel = ["a", "e", "i" ,"o" ,"u"]
print(vowel[0:3]) #output = ['a', 'e', 'i']
print(vowel[3:5]) #output = ['o', 'u']
```
**List Methods**
* `append()` Method add an item to the end of the list.
* `insert()` Method add an item to a specific index.
* `pop()` Method remove an item from a list based on the value.
* `remove()` Method remove an item at a particular index.
* `clear()` Remove all item from the list.
* `copy()` Return a shallow copy of the list.
* `count()` Return the number of time a value appeared in the list.
* `extend()` Appends another list to the current list by expending it.
* `index()`  Return the index of the value inside it.
* `reverse()` Reverse the list in place.
* `sort()` Sorts the list in place.

**Example**
```py
Book = ["Harry Potter", "1984", "The Fault in Our Stars", "The Mom Test", "Life in Code"]
Book.append("Pachinko")
Book.remove("1984")
Book.pop(1)
print(Book)
```
