* The `pwd` command tells you your current location (which directory you are in).
* The `cat` command can be used to view `.txt` files or `.py` files directly from the terminal.
* You can use `cd ..` to go back one directory.
* `cd ../..` moves back multiple directories.
* The `ls` command is used to see the contents of a directory.
* `echo` is used to print anything you want.
* The `clear` command is used to clear the terminal when it gets messy.
* Multiple keyboard shortcuts can be used in the terminal, like arrow keys and `Tab`.
* `cd` is used to change the directory you are working in.

**File and Directory Management Commands**

**Creating Files and Directories**
- **`mkdir {new-directory}`** - Creates a new directory
- **`touch {new-file}`** - Creates a new empty file

**Writing to Files**
- **`echo "text" > file.txt`** - Creates/overwrites file with text
- **`echo "text" >> file.txt`** - Appends text to file

**Reading Files**
- **`cat file.txt`** - Displays file contents

**Combining Files**
- **`cat file-a > file-b`** - Copies file-a to file-b (overwrites)
- **`cat file-a >> file-b`** - Appends file-a to file-b
- **`cat file-a file-b > file-c`** - Combines two files into file-c
- **`cat file-a file-b file-c > file-d`** - Combines three files into file-d

**Removing Files and Directories**
- **`rm file.txt`** - Removes a file
- **`rm file1.txt file2.txt`** - Removes multiple files
- **`rm -r directory/`** - Recursively removes directory and contents
- **`rm -i file.txt`** - Interactive mode (prompts before deletion)
- **`rm -f file.txt`** - Forces deletion without prompting
- **`rmdir directory/`** - Removes empty directory only

**Copying Files and Directories**
- **`cp file.txt file-copy.txt`** - Copies a file
- **`cp file1.txt file2.txt directory/`** - Copies multiple files to directory
- **`cp -r directory/ directory-copy/`** - Recursively copies directory and contents
- **`cp -i file.txt file-copy.txt`** - Interactive mode (prompts before overwriting)

**Listing Directory Contents**
- **`ls -R`** - Recursively lists all files and subdirectories in tree format
