# 🐍 Python Built-in Functions Guide

Welcome to the quick reference guide for Python's most commonly used built-in functions. To make learning easier, they are grouped by their primary purpose.

### 🔢 Math & Numbers
Functions used for mathematical operations and calculations.

| Function | Description | Example | Output |
| :--- | :--- | :--- | :--- |
| `abs()` | Returns the absolute (positive) value of a number. | `abs(-5)` | `5` |
| `max()` | Returns the largest item from an iterable or multiple arguments. | `max(1, 10, 5)` | `10` |
| `min()` | Returns the smallest item from an iterable or multiple arguments. | `min(1, 10, 5)` | `1` |
| `round()` | Rounds a number to a specified number of decimal places. | `round(3.14159, 2)` | `3.14` |
| `sum()` | Adds together all items of an iterable (like a list). | `sum([1, 2, 3])` | `6` |

### 🔄 Type Conversions
Functions used to convert data from one type to another.

| Function | Description | Example | Output |
| :--- | :--- | :--- | :--- |
| `bool()` | Converts a value to a boolean (`True` or `False`). | `bool(1)` | `True` |
| `dict()` | Creates a dictionary (key-value pairs). | `dict(a=1, b=2)` | `{'a': 1, 'b': 2}` |
| `float()` | Converts a value to a floating-point (decimal) number. | `float(5)` | `5.0` |
| `int()` | Converts a value to an integer (whole number). | `int(5.9)` | `5` |
| `list()` | Creates a list from an iterable. | `list("cat")` | `['c', 'a', 't']` |
| `set()` | Creates a set (a collection of unique items). | `set([1, 2, 2, 3])` | `{1, 2, 3}` |
| `str()` | Converts a value into a string (text). | `str(100)` | `'100'` |

### 📦 Iterables & Collections
Functions that help you work with sequences like lists, strings, and tuples.

| Function | Description | Example | Output |
| :--- | :--- | :--- | :--- |
| `all()` | Returns `True` if **all** items in an iterable are true. | `all([True, 1, "yes"])` | `True` |
| `any()` | Returns `True` if **any** item in an iterable is true. | `any([False, 0, "yes"])`| `True` |
| `enumerate()`| Adds a counter to an iterable, returning `(index, item)` pairs. | `list(enumerate(['a', 'b']))`| `[(0, 'a'), (1, 'b')]`|
| `filter()`| Keeps only the items that return `True` for a given function. | `list(filter(lambda x: x>0, [-1, 2]))`| `[2]` |
| `len()` | Returns the length (total number of items) of an object. | `len("hello")` | `5` |
| `map()` | Applies a specific function to every item in an iterable. | `list(map(len, ['hi', 'cat']))`| `[2, 3]` |
| `range()` | Generates a sequence of numbers (great for loops). | `list(range(3))` | `[0, 1, 2]` |
| `zip()` | Combines multiple iterables element-by-element. | `list(zip([1, 2], ['a', 'b']))`| `[(1, 'a'), (2, 'b')]`|

### 🛠️ Utilities & Helpers
Functions that help you debug, inspect, and display data.

| Function | Description | Example | Output |
| :--- | :--- | :--- | :--- |
| `dir()` | Returns a list of properties and methods associated with an object. | `dir([])` | `['append', 'clear', ...]` |
| `print()` | Prints output to the console for the user to see. | `print("Hello!")` | `Hello!` |
| `type()` | Tells you what data type an object is. | `type(3.14)` | `<class 'float'>` |
